```python
q1 = my_data["Sales"].quantile(.25)
q3 = my_data["Sales"].quantile(.75)
iqr = q3 - q1
print(q1, q3, iqr)
                    
```

```python
#calculating the lower bound
lower_bound = q1 - 1.5*iqr
```

```python
lower_bound
```

```python
#calculating the upper bound of data
upper_bound = q3 + 1.5*iqr
```

```python
upper_bound
```

```python
required = my_data[(my_data["Sales"] >= lower_bound) & (my_data["Sales"] <= upper_bound)]
print ("All sales\n", my_data["Sales"].values)
print(" ")
print("Sales after removing outliers\n", required["Sales"].values)
```
