### GROUPING AND AGGREGATION
We want to group data by the department


```python
df
```

```python
newdata = df.groupby("department")
```

```python
newdata 
```

```python
newdata = df.groupby("department")["salary"].sum()
```

```python
newdata= df.groupby("department")["salary"].sum()
```

```python
newdata
```

```python
newdata1 = df.groupby("department")["salary"].sum()/3
```

```python
newdata1
```

```python
newdata1 = df.groupby("department")["salary"].mean() #finds the mean
```

```python
newdata1
```

```python
#groups and calculates all the values using the functions
df.groupby("department").agg({"salary":["sum", 'mean', 'count','max', 'std','var','median'],})
```

```python
df.groupby("department").agg({"age":["sum", 'mean', 'count','max', 'std','var','median'],})
```
